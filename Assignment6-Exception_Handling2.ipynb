
Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

```python
'''Ans 4.

LookupError is used to represent a category of exceptions that occur when lookup operations fail or encounter issues, 
where lookup operations represent searching or accessing elements and items in data structures like lists, dictionaries,
or sets. Below are some EookupErrors

IndexError

KeyError

NameError

'''

#KeyError
try:
    d={"1": "English", "2": "French"}
    d["3"]
except KeyError as e:
    print("KeyError \n")
    
#IndexError
try:
    l=[1,2,3]
    l[5]
except IndexError as e:
    print(e)
```


Q6. List down some best practices for exception handling in python. 

Ans 6

Some practices for exception handling in Python are as follows:

1. Be Specific: Catch exceptions by their specific types for precise error handling.
2. Avoid Bare except: Avoid catching all exceptions with a bare except, use specific exceptions or a base exception like Exception.
3. Use finally: Use finally for cleanup code that must run, regardless of whether an exception occurs.
4. Raise When Needed: Raise exceptions when your code encounters errors or invalid states.
5. Logging: Use logging to record error information instead of printing to the console.

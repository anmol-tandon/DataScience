Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in
Jupyter Notebook.

```python
#A2. 

from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "Hello, World!"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

Q3. What is App routing in Flask? Why do we use app routes?

A3. App routing in Flask refers to the process of defining URL routes and mapping them to specific view functions within a Flask application.

App routes are used to handle incoming requests and decide what content or functionality to present based on the requested URL. It allows us to determine how your application responds to different URLs and HTTP methods (e.g., GET, POST, PUT, DELETE).
Also through app routes, we can access the server globally and with app routes server system becomes language independent which mean we can access it no matter code is written in any language.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

```python
'''A5.

In Flask, the url_for() function is used for URL building. With this, we can dynamically create URLs for different 
parts of our Flask application, making it easier to maintain and update your application's URL structure.
'''

from flask import Flask, url_for
app = Flask(__name__)

# Define two routes
@app.route('/')
def home():
    return 'Welcome to the Home Page'

@app.route('/about')
def about():
    return 'This is the About Page'

if __name__ == '__main__':
    with app.test_request_context():
        # Use url_for() to generate URLs for the 'home' and 'about' routes
        home_url = url_for('home')
        about_url = url_for('about')
        
        # Print the generated URLs
        print("URL for the Home Page:", home_url)
        print("URL for the About Page:", about_url)

    app.run(host="0.0.0.0")

```
